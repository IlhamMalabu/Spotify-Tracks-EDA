```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Importing Track Data

```python
df_tracks = pd.read_csv("/Users/zainabumalabu/Downloads/archive(1)/tracks.csv")
df_tracks.head()
```

## Checking for Null Values in Data

```python
#check for null values
pd.isnull(df_tracks).sum()
```

```python
#check for total num of rows and columns & datatypes
df_tracks.info()
```

## Least Popular Songs in Dataset

```python
#find the least popular songs in dataset
least_popular = df_tracks.sort_values('popularity', ascending = True).head(10)
least_popular
```

## Most Popular Songs in Dataset

## Descriptive Stats for numerical variables

```python
#descriptive stats for the numerical variables
df_tracks.describe().transpose()
```

```python
#setting release date as index
df_tracks.set_index('release_date', inplace = True)
df_tracks.index = pd.to_datetime(df_tracks.index)
df_tracks.head()
```

```python
#filtering specific rows
#df_tracks[['artists']].iloc[18]
```

```python
#converting duration in ms to s
df_tracks['duration'] = df_tracks['duration_ms'].apply(lambda x: round(x/1000))
df_tracks.drop('duration_ms', inplace = True, axis = 1)
df_tracks.duration.head()
```

## Correlation Map

```python
#correlation map
corr_df = df_tracks.drop(['key', 'mode', 'explicit'], axis = 1).corr(method = "pearson")
plt.figure(figsize = (14,6))
heatmap = sns.heatmap(corr_df, annot = True, fmt = '.1g', vmin=-1, center=0, cmap="inferno", linewidths = 1, linecolor = "Black")
heatmap.set_title("Correlation HeatMap Between Variable")
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation = 90)
```

```python
#sample
sample_df = df_tracks.sample(int(0.004 * len(df_tracks)))
```

```python
print(len(sample_df))
```

## Correlation between loudness and energy

```python
#correlation between loudness and energy
plt.figure(figsize = (10,6))
sns.regplot(data = sample_df, y = 'loudness', x = 'energy', color = 'c').set(title = 'loudness vs energy Correlation')
#There is a positive correlation between the loudness and correlation 
```

## Correlation between popularity and acousticness

```python
plt.figure(figsize = (10,6))
sns.regplot(data = sample_df, y = 'popularity', x = 'acousticness', color = 'c').set(title = 'popularity vs acousticness Correlation')
#downward trend of regression line implies that the more acoustic the less popular
```

```python
#changing release date format
df_tracks['dates'] = df_tracks.index.get_level_values('release_date')
df_tracks.dates = pd.to_datetime(df_tracks.dates)
years = df_tracks.dates.dt.year
```

## Distribution plot to visualize total number of songs in each year

```python
#distribution plot to visualize total number of songs in each year
sns.displot(years, discrete = True, aspect = 2, height = 5, kind = 'hist').set(title = "Number of songs per year")
```

```python
#duration of songs over the years
total_dr = df_tracks.duration
fig_dims = (18,7)
fig, ax = plt.subplots(figsize = fig_dims)
fig = sns.barplot(x = years, y = total_dr, ax = ax, errwidth = False).set(title = "Year vs Duration")
plt.xticks(rotation = 90)
```
